# Neural Style Transfer

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import functools

import tensorflow as tf
import tensorflow.contrib.eager as tfe
from tensorflow.python.keras import models, layers, losses
from tensorflow.python.keras import backend as K
from tensorflow.keras.preprocessing import image as kp_image
```

```python
tf.enable_eager_execution()
print (tf.executing_eagerly())
```

```python
# Paths to content and style images
content_path = './harini.JPG'
style_path = './kanagawa.jpg'
```

```python
def load_img(path_to_img):
    """
    Input: Path to image.
    Output: Resized image (compatible with VGG-19).
    """
    max_dim = 512
    img = Image.open(path_to_img)
    long = max(img.size)
    scale = max_dim/long
    #Resizing img to use with VGG
    img = img.resize((round(img.size[0]*scale), round(img.size[1]*scale)), Image.ANTIALIAS)

    img = kp_image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    return img
```

```python
def imshow(img, title=None):
    """
    Input: image.
    Normalizes and displays the image.
    """
    # Remove the batch dimension
    if len(img.shape) ==4:
        out = np.squeeze(img, axis=0)
    else:
        out = img
    # Normalize for display 
    out = out.astype('uint8')
    plt.imshow(out)
    if title is not None:
        plt.title(title)
    plt.imshow(out)
```

```python
# Display the content image
plt.figure(figsize=(10,10))
content = load_img(content_path).astype('uint8')
plt.subplot(1,2,1)
imshow(content)
plt.title ('Content image')

# Display the style image
style = load_img(style_path).astype('uint8')
plt.subplot(1,2,2)
imshow(style)
plt.title ('Style image')
```

```python
def load_and_process_img(path_to_img):
    """
    Input: Path to image.
    Output: Preprocessed image Using vgg-19's preprocess_input().
    """
    image = load_img(path_to_img)
    image = tf.keras.applications.vgg19.preprocess_input(image)
    return image
```

```python
# Display the preprocessed content image
plt.figure(figsize=(10,10))
content = load_and_process_img(content_path).astype('uint8')
plt.subplot(1,2,1)
imshow(content)
plt.title ('Content image')

# Display the preprocessed style image
style = load_and_process_img(style_path).astype('uint8')
plt.subplot(1,2,2)
imshow(style)
plt.title ('Style image')
```

```python
def deprocess_img(processed_img):
    """
    Input: processed image.
    Output: Returns original image (without processing).
    """
    x = processed_img.copy()
    if len(x.shape) == 4:
        x = np.squeeze(x, 0)
    assert len(x.shape) == 3, ("Input to deprocess image must be an image of "
                             "dimension [1, height, width, channel] or [height, width, channel]")
    if len(x.shape) != 3:
        raise ValueError("Invalid input to deprocessing image")

    # perform the inverse of the preprocessing step
    x[:, :, 0] += 103.939
    x[:, :, 1] += 116.779
    x[:, :, 2] += 123.68
    x = x[:, :, ::-1]

    x = np.clip(x, 0, 255).astype('uint8')
    return x
```

```python
# Display content image that was reverse-preprocessed
plt.figure(figsize=(10,10))
content = deprocess_img(load_and_process_img(content_path))
plt.subplot(1,2,1)
imshow(content)
plt.title ('Content image')

# Display style image that was reverse-preprocessed
style = deprocess_img(load_and_process_img(style_path))
plt.subplot(1,2,2)
imshow(style)
plt.title ('Style image')
```

```python
vgg = tf.keras.applications.vgg19.VGG19()
vgg.summary()
```

```python
# Define style and content layers
content_layers = ['block5_conv2'] #later layers
style_layers = ['block1_conv1', 'block2_conv1', 'block3_conv1', 'block4_conv1', 'block5_conv1'] #initial layers
```

```python
def get_model():
    """
    Output: Returns the model input and features from style and content layers.
    """
    vgg = tf.keras.applications.vgg19.VGG19(include_top=False, weights='imagenet')
    vgg.trainable = False
    content_outputs = [vgg.get_layer(i).output for i in content_layers]
    style_outputs = [vgg.get_layer(i).output for i in style_layers]
    
    model_outputs = style_outputs + content_outputs
    return models.Model(vgg.input, model_outputs)
```

```python
def get_content_loss(gen_content, target_content):
    """
    Input: Content features for generated and target images.
    Output: Returns the content loss.
    """
    return tf.reduce_mean(tf.square(gen_content - target_content))
```

```python
def gram_matrix(x):
    """
    Input: matrix.
    Output: gram matrix.
    """
    channels = int(tf.shape(x)[-1]) #eg:28*28*32 --> we need 32
    x_v = tf.reshape(x, [-1,channels]) #-1 --> multiplies 28*28 and returns 28*28,32
    n_entries = tf.shape(x_v)[0]
    
    G = tf.matmul(x_v, x_v, transpose_a=True)
    G = G / tf.cast(n_entries, tf.float32)
    
    return G

def get_style_loss(gen_style, G_target_style):
    G_gen_style = gram_matrix(gen_style)
    return tf.reduce_mean(tf.square(G_gen_style - G_target_style))
```

```python
def get_features(model, content_path, style_path):
    """
    Input: model, content image path, style image path.
    Output: Returns content image and style image features.
    """    
    content_img = load_and_process_img(content_path)
    style_img = load_and_process_img(style_path)
    
    content_img_features = model(content_img)
    style_img_features = model(style_img)
    
    content_img_features_filt = [layer[0] for layer in content_img_features[5:]]
    style_img_features_filt = [layer[0] for layer in style_img_features[:5]]

    return style_img_features_filt, content_img_features_filt
```

```python
def compute_loss(model, alpha, beta, gen_image, G_target_style, target_content):
    """
    Input: model, content weight (alpha), style weight (beta), generated image, 
           target style image gram matrix, target content image.
    Output: loss, weighted style score, weighted content score.
    """
    
    model_outputs = model(gen_image)
    gen_style = model_outputs[:5]
    gen_content = model_outputs[5:]
    
    style_score = 0
    content_score = 0
    
    wt_per_layer = 1.0/5.0
    for inp1,inp2 in zip(gen_style, G_target_style):
        style_score += get_style_loss(inp1,inp2) *wt_per_layer
    
    wt_per_layer = 1.0
    for inp1,inp2 in zip(gen_content, target_content):
        content_score += get_content_loss(inp1,inp2) *wt_per_layer
    
    
    loss = alpha*content_score + beta*style_score
    
    return loss, beta*style_score, alpha*content_score
```

```python
def compute_grads(fg):
    """
    Input: model, content weight (alpha), style weight (beta), generated image, 
           target style image gram matrix, target content image.
    Output: gradient, loss.
    """
    with tf.GradientTape() as tape: 
        all_loss = compute_loss(**fg)
    total_loss = all_loss[0]
    return tape.gradient(total_loss, fg['gen_image']), all_loss
```

```python
import IPython.display

# alpha more important than beta because content more important
def style_transfer_main(content_path, style_path, nIters=1000, alpha=1e3, beta=1e-2):
    """
    Input: path to content image, path to style image, number of iterations, 
           content weight, style weight
    Output: best image, best loss
    """
    
    #Step1: Create model to produce features from style/content layers
    model = get_model()
    
    # Layers are unchangable
    for layer in model.layers:
        layer.trainable = False
    
    #Step2: Extract style/content features for the target images
    target_style, target_content = get_features(model, content_path, style_path)
    G_target_style = [gram_matrix(x) for x in target_style]

    #Step3: Initialize gen_image
    init_img = load_and_process_img(content_path)
    gen_image = tf.Variable(init_img, dtype=tf.float32)
    
    opt = tf.train.AdamOptimizer(learning_rate=5, beta1=0.99, epsilon=1e-1)
    
    best_loss = 1e50
    best_img = np.zeros(init_img.shape)
    
    fg = {'model':model, 'alpha':alpha, 'beta':beta, 'gen_image':gen_image, 
          'G_target_style':G_target_style, 'target_content':target_content}
    
    ref_means = np.array([103.939, 116.779, 123.68])
    min_vals = -ref_means
    max_vals = 255 - ref_means
    imgs = []
    
    for i in range(nIters):
        grads, all_loss = compute_grads(fg)
        loss, style_loss, content_loss = all_loss
        opt.apply_gradients([(grads, gen_image)])
        clip = tf.clip_by_value(gen_image, min_vals, max_vals)
        
        gen_image.assign(clip)
        
        if loss < best_loss:
            best_loss = loss
            best_img = deprocess_img(gen_image.numpy())


        print ('Iteration ',i, ': Losses = ', format(loss), format(style_loss), format(content_loss) )
        if (i+1)%100 == 0:
            #print ('Iteration ',i, ': Losses = ', loss, style_loss, content_loss )
            imgs.append(best_img)

    if len(imgs)==10:
        it = 0
        fig,ax = plt.subplots(2,5,figsize=(14,6))
        for row in range(2):
            for col in range(5):
                ax[row,col].imshow(imgs[it])
                ax[row,col].set_xticks([])
                ax[row,col].set_yticks([])
                it += 1

    return best_img, best_loss
```

```python
best_img, best_loss = style_transfer_main(content_path, style_path, nIters=1000)
```
